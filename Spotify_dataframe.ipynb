```python
import pandas as pd
```

```python
file_names = ['big.csv', 'charts_2023.csv', 'partyhits.csv', 'pop2023.csv', 'popular.csv', 
              'rock_playlist.csv', 'tiktok.csv', 'tophits.csv', 'big1.csv', 'big2.csv']

spotify_df = []

for filename in file_names:
    df = pd.read_csv(filename)
    spotify_df.append(df)

spotify_df = pd.concat(spotify_df, ignore_index=True)
spotify_df
```

```python
# drop the columns 
spotify_df.drop(['Unnamed: 0', 'key', 'mode', 'valence', 'type', 'uri', 
                    'track_href', 'analysis_url', 'duration_ms', 'time_signature'], axis= 1, inplace= True)
```

```python
spotify_df.to_csv('audiofeatures.csv')
```
